```python
import requests
import folium
import datetime
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import gmplot
from IPython.display import IFrame
 

def location_coordinates():
    try:
        response = requests.get('https://ipinfo.io')
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()

        loc = data['loc'].split(',')
        lat, long = float(loc[0]), float(loc[1])
        city = data.get('city', 'Unknown')
        state = data.get('region', 'Unknown')

        return lat, long, city, state

    except requests.RequestException as e:
        print(f"Error fetching location data: {e}")
        return None

def create_map(lat, long, city, state):
    try:
        map_obj = folium.Map(location=[lat, long], zoom_start=12)
        folium.Marker([lat, long], popup=f'Current Location: {city}, {state}').add_to(map_obj)

        file_name = f"Location_{datetime.date.today()}.html"
        map_obj.save(file_name)
        return file_name

    except Exception as e:
        print(f"Error creating map: {e}")
        return None

def open_map(file_name):
    try:
        # Setup the Chrome WebDriver
        service = Service(ChromeDriverManager().install())
        dr = webdriver.Chrome(service=service)
        
        # Open the map file
        dr.get(f"file://{IFrame(src='my_map.html', width=700, height=600)}")
        

        time.sleep(4)
        dr.quit()
        print("\nBrowser Closed..........")

    except Exception as e:
        print(f"Error opening map file in browser: {e}")

if __name__ == "__main__":
    print("---------------GPS Using Python---------------\n")
    
    coordinates = location_coordinates()
    if coordinates:
        lat, long, city, state = coordinates
        print(f"You Are in {city}, {state}")
        print(f"Your latitude = {lat} and longitude = {long}")

        map_file = create_map(lat, long, city, state)
        if map_file:
            print(f"\nOpening File: {map_file}")
            open_map(map_file)
        else:
            print("Failed to create the map.")
    else:
        print("Failed to get location coordinates.")
```
